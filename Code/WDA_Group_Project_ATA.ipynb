```python
  
def process_dollar_data(temp):
    # Initialize index variable
    i = 1
    # Loop through and remove every second element starting from index 1 (the '$' signs)
    while i < len(temp):
        temp.pop(i)  # Remove the '$' sign
        i += 1  # Move to the next value

    # Append '$' to each remaining value
    for i in range(1, len(temp)):
        temp[i] = '$' + temp[i]
    
    return temp
    
def process_percentage_data(temp):
    # Initialize index variable
    i = 1
    # Loop through and remove every second element starting from index 1
    while i < len(temp):
        temp.pop(i + 1)
        i += 1
    # print(temp)
    # Append '%' to each element from index 1 onwards
    for i in range(1, len(temp)):
        if ('(') in temp[i]:
            temp[i]+=')'
        temp[i] = temp[i] + '%'
    
    return temp

def set_first_row_as_header(data):
    df = pd.DataFrame(data)
    df.columns = df.iloc[0]  # Set the first row as header
    df = df[1:]  # Remove the first row
    df.reset_index(drop=True, inplace=True)
    return df
```
